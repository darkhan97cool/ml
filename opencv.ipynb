```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt



#1
img = cv.imread('dk.jpg', 10)
plt.imshow(img,cmap='gray')
plt.show()

#2
img1 = cv.imread('dk.jpg', cv.IMREAD_GRAYSCALE)
resized_image = cv.resize(img1, (1400, 900)) 
plt.imshow(resized_image, cmap='gray')
plt.show()

#3
(h, w) = resized_image.shape[:2]
center = (w / 2, h / 2)
  
scale = 1.0

M = cv.getRotationMatrix2D(center, 295, scale)
rotated270 = cv.warpAffine(resized_image, M, (w, h))

plt.imshow(rotated270, 'gray')
plt.show()
```
